```python
#put into a pandas dataframe

trinidad_temp_df = pd.read_csv('Trinidad_TEMP_1973-201905.txt', skiprows=32, sep="\t", header=None)

# trinidad_temp_df = pd.DataFrame(lol)
trinidad_temp_df
```

```python
new_trinidad_temp_df = trinidad_temp_df[[0, 1, 2, 3, 4, 5, 6, 7, 8]].copy()
```

```python
new_trinidad_temp_df
```

```python
new_trinidad_temp_df = new_trinidad_temp_df.rename(columns={0: "Year",
                                                          1: "Month",
                                                            2: "Day",
                                                            3: "TIME_PST",
                                                            4: "TIME_FLAG",
                                                            5: "TRINIDAD_BAY_SURF_TEMP_C",
                                                            6: "BAY_FLAG",
                                                            7: "TRINIDAD_BEACH_SURF_TEMP_C",
                                                            8: "BEACH_FLAG"})
```

```python
new_trinidad_temp_df
```

```python
trin_df = new_trinidad_temp_df.dropna(axis=0,how='any', thresh=None, subset=None, inplace=False)
```
